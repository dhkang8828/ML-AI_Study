### Activation Functions

```python
import torch
import torch.nn as nn

from matplotlib import pyplot as plt
```

```python
## randn : 정규분포(평균0, 편차1)에 따라 100개의 sample (*10을 했기 때문에 평균0, 편차10)
## sort[0]: sorting된 tensor
## sort[1]: indices
x = torch.sort(torch.randn(100) * 10)[0]
print(f"x:{x}")
```

### Sigmoid

```python
act = nn.Sigmoid()
print(act(x))
```

```python
print(f"torch.sigmoid: {torch.sigmoid(x)}")
```

```python
plt.plot(x.numpy(), torch.sigmoid(x).numpy())
plt.show()
```

### Hypoerbolic Tangent(TanH)

```python
act = nn.Tanh()
print(act(x))
```

```python
print(f"torch.tanh(x): {torch.tanh(x)}")
```

```python
plt.plot(x.numpy(), torch.tanh(x).numpy())
plt.show()
```
